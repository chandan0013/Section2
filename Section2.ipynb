```python
import pandas as pd
df = pd.read_table('C:/Users/Chandan/Desktop/DATA/2017.txt',index_col=False)
df1 = pd.read_table('C:/Users/Chandan/Desktop/DATA/2016.txt',index_col=False)
df
```

```python
## NAN considered as 0 entry, so average n arithmatic computation consider only non-suppresed data
##Question 1: average number of beneficiaries per provider
total_rows = len(df.index)
summ = df['bene_count'].sum()
avg = summ/total_rows
avg

##Question 2: median, in days, of the distribution of this value across all providers
val1 = df["total_day_supply"] / df["total_claim_count"]
val1.median()

##Question 3: Fraction of drugs for brand name
tot_claims = df["total_claim_count"].to_numpy()
sum1 = df['total_claim_count'].sum()
sum2 = 0
for i in range(len(tot_claims)):
    if tot_claims[i] >= 1000:
        sum2 = sum2 + tot_claims[i]
frac = sum2/sum1
        

##Question 4: Ratio of opiniod to antibiotics beneficiaries
states = df['nppes_provider_state'].to_numpy()
uniqueStates = df['nppes_provider_state'].unique()
opiods = df['opioid_claim_count'].to_numpy()
antibio = df['antibiotic_claim_count'].to_numpy()
val2 = []
for i in range(len(uniqueStates)):
    sum1 = 0
    sum2 = 0
    for j in range(len(states)):
        if uniqueStates[i] == states[j]:
            sum1 = sum1 + opiods[j]
            sum2 = sum2 + antibio[j]
    val2.append(sum1/sum2)
sol = max(val2) - min(val2)
            
##Question 5: Pearson correlation
val3 = df["bene_count_ge65"]
val4 = df["lis_claim_count"]
val3.corr(val4)

##Question 7: Average Inflation Rate
val5 = df["total_drug_cost"] / df["total_day_supply"]
val6 = df1["total_drug_cost"] / df1["total_day_supply"]
inflation_rate = val5 - val6
avg_inflation_rate = inflation_rate.sum()/total_rows

##Question 8:
Spec1 = df1["specialty_description"].to_numpy()
Spec = df["specialty_description"].to_numpy()
uniqueSpec = df1["specialty_description"].unique()
providers_2016 = []
providers_2017 = []
for i in range(len(uniqueSpec)):
    sum1 = 0
    sum2 = 0
    for j in range(len(Spec1)):
        if uniqueSpec[i] == Spec1[j]:
            sum1 = sum1 + 1;
            
    providers_2016.append(sum1)
    
    for j in range(len(Spec)):
        if uniqueSpec[i] == Spec[j]:
            sum2 = sum2 + 1;
    
    providers_2017.append(sum2)
    
frac1 = []
for i in range(len(uniqueSpec)):
    if providers_2016[i] >= 1000:
        frac1.append((providers_2017[i] - providers_2016[i]) / len(uniqueSpec))

max(frac1)
```
